```python
import plotly as py
import plotly.graph_objs as go
```

```python
%matplotlib inline
#%%pprint
```

```python
from plotly.offline import download_plotlyjs
```

```python
from plotly.offline import init_notebook_mode
from plotly.offline import plot, iplot
```

```python
init_notebook_mode(connected =True)
```

```python
mydata =dict(type='choropleth', locations =['NY','AZ','AL','CA','AK','AR'], locationmode='USA-states',
            colorscale='Greens', text=['Food','Water','Hospital','Flood','schools','roads'], z =[1.0,2.0,3.0,890,654,873], 
            colorbar ={'title':'Colorbar title of the plot'})
```

```python
#mydata
```

```python
mylayout =dict(geo={'scope':'usa'})
```

```python
choromap =go.Figure(mydata, mylayout)
```

```python
iplot(choromap)
```

## Using plot will plot the map as an html file which you can easily plot

```python
plot(choromap)
```
