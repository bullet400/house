```python
prices =df['PRICE']
features =df.drop('PRICE', axis =1)
```

```python
print(prices)
```

```python
features
```

## Training and testing our model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test =train_test_split(features, prices, test_size=0.2, random_state =2)
```

## run our regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor =LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

## Performing our prediction of prices

```python
y_predict =regressor.predict(X_test)
```

```python
y_predict
```

```python
print('The intercept is', regressor.intercept_)
```

```python
print('The coefficient is',regressor.coef_)
```

```python
total_regressor = pd.DataFrame(data =regressor.coef_, index=X_train.columns, columns=['Coef'])
```

```python
total_regressor
```

###### Obtaining R-Square R-squared is a goodness-of-fit measure for linear regression models.

```python
print('The r-square is',regressor.score(X_test, y_test))
```
